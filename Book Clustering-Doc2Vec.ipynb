Data clean

```
lst_stopwords = nltk.corpus.stopwords.words("english")
def utils_preprocess_text(text):
    ## clean (convert to lowercase and remove punctuations and characters and then strip)
    text = re.sub(r'[^\w\s]', ' ', str(text).lower().strip())
            
    ## Tokenize (convert from string to list)
    lst_text = text.split()
    lst_text_long = []
    for i in lst_text:
      if len(i) > 3:
        lst_text_long.append(i)
    ## remove Stopwords
    lst_text_stop = [word for word in lst_text_long if word not in 
                    lst_stopwords]
    #print(lst_text_stop)                      
    ## Lemmatisation (convert the word into root word)
    lem = nltk.stem.wordnet.WordNetLemmatizer()
    lst_text_lem = [lem.lemmatize(word) for word in lst_text_stop]
    #print(lst_text_lem)        
    ## back to string from list
    text = " ".join(lst_text_lem)
    return text
```

```
#Define "chunks" and use it to divide each book into different chunks. 
def chunks(text_split,n):
    for i in range(0, len(text_split),n):
        if i//n<len(text_split)//n:
          yield ' '.join(text_split[i:i + n])
random.seed(0)

#Each chunk except for the last one has 100 words.Randomly select 200 chunks for each book, except the last chuck

n_words=150
n_chuncks=200

text1 = list(chunks(book1_tokenized,n_words))
text_1 = random.sample(text1[0:len(text1)],n_chuncks)
text2 = list(chunks(book2_tokenized,n_words))
text_2 = random.sample(text2[0:len(text2)],n_chuncks)
text3 = list(chunks(book3_tokenized,n_words))
text_3 = random.sample(text3[0:len(text3)],n_chuncks)
text4 = list(chunks(book4_tokenized,n_words))
text_4 = random.sample(text4[0:len(text4)],n_chuncks)
text5 = list(chunks(book5_tokenized,n_words))
text_5 = random.sample(text5[0:len(text5)],n_chuncks)

print(len(book1_tokenized)//n_words, len(book2_tokenized)//n_words,len(book3_tokenized)//n_words,len(book4_tokenized)//n_words,len(book5_tokenized)//n_words)
```

```
import gensim
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
class Doc2VecModel(BaseEstimator, TransformerMixin):
  def __init__(self,min_count=1,window=2,vector_size=100,workers=4):
    super().__init__()
    self.min_count=min_count
    self.window=window
    self.vector_size=vector_size
    self.workers=workers
    self.model=Doc2Vec(min_count=self.min_count, window=self.window, vector_size=self.vector_size, workers=self.workers)

  def fit(self,raw_documents,y=None):
    return self

  def transform(self,raw_documents, y=None):
    #raw_documents=dtf['sample']
    docs=[]
    for i in range(len(raw_documents)):
      docs.append(TaggedDocument(words=raw_documents[i].split(), tags=[i]))
    model_doc=self.model
    model_doc.build_vocab(docs)
    model_doc.train(docs, total_examples=model_doc.corpus_count, epochs=model_doc.iter)
    vec_list=[]
    for i in range(len(raw_documents)):
      vector=model_doc.infer_vector(raw_documents[i].split())
      vec_list.append(vector)
    return vec_list

vec=Doc2VecModel()
#vec_=vec.transform(dtf['sample'])
#print(vec_)
```

```
#Evaluation kappa?:κappa的值在−1到1之间。 越接近1，两者越一致、越吻合。接近0，表面两者之间符合偶然的预期。接近-1，表面两者之间的相符程度非常低。
#contingency matrix
import numpy as np
from sklearn.metrics.cluster import contingency_matrix
print(set(dtf2['label']))
comparision_label=['label_kmeans','label_em','label_Hier']

def kappa(label_i):
  matrix=contingency_matrix(dtf2['label'],dtf2[label_i])
  print(f'matrix with {label_i}')
  print(matrix)
  b3=np.zeros((1,1))
  if matrix.shape[1]!=5:
    for i in range(5-matrix.shape[1]):
      matrix=np.insert(matrix,matrix.shape[1],values=b3,axis=1)
  pe_rows=np.sum(matrix,axis=0)
  pe_cols=np.sum(matrix,axis=1)
  sum_total=sum(pe_cols)
  pe=np.dot(pe_rows, pe_cols)/float(sum_total**2)
  po=np.trace(matrix)/float(sum_total)
  print(f'kappa with {label_i} is {(po-pe)/(1-pe)}') 
  print(f'acc with {label_i} is {po}')
  return (po-pe)/(1-pe)

for label_i in comparision_label:
  _=kappa(label_i)
```

```
#Coherence?
#Rand index: measures the similarity of the two assignments, ignoring permutations; rand_score/adjusted_rand_score
#The adjusted Rand index is thus ensured to have a value close to 0.0 for random labeling independently of the number of clusters and samples and exactly 1.0 when the clusterings are identical (up to a permutation).
#Mutual Information based scores:measures the agreement of the two assignments,ignoring permutations.
#mutual_info_score, adjusted_mutual_info_score and normalized_mutual_info_score 
from sklearn import metrics
def adjusted_rand_score(label_i):
  score=metrics.adjusted_rand_score(dtf2['label'],dtf2[label_i])
  print(f'adjusted_rand_score with {label_i} is {score}')
  return score
def adjusted_mutual_info_score(label_i):
  score=metrics.adjusted_mutual_info_score(dtf2['label'],dtf2[label_i])
  print(f'adjusted_mutual_info_score with {label_i} is {score}')
  return score

for label_i in comparision_label:
  _=adjusted_rand_score(label_i)
print('\n')
for label_i in comparision_label:
  _=adjusted_mutual_info_score(label_i)
```

```
#Evaluation
#Calculate Kappa against true authors, Consistency , Coherence and Silhouette
#Silhouette:[-1,1]; The score is higher when clusters are dense and well separated. Scores around zero indicate overlapping clusters.
from sklearn import metrics

def Silhouette(label_i):
  silhouette=metrics.silhouette_score(data_test,dtf2[label_i],metric='euclidean')
  print(f'Silhouette of {label_i}:{silhouette}')
  return silhouette

for label_i in comparision_label:
  _=Silhouette(label_i)
```
